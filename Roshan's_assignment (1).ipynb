## Objective

The objective of this project is to create a simple calculator using Python.
This calulator perfroms basic arthimatic calculations like addition, subtraction, division,
multiplication using contitional statements.

```python
# Taking input form the user
a = int(input("Enter the first number: "))
b = int(input("Enter the second number: "))
operator = input("Choose an operator (+,-,*,/): ")

# using conditional statements
if operator == "+":
    result = a + b
    print("result",result)
elif operator == "-":
    result = a - b
    print("result:",result)
elif operator == "*":
    result = a * b
    print("result:",result)
elif operator == "/":
    if b == 0:
        print("error: division by zero is not allowed!")
    else:
        result == a / b
        print("result:",result)
else:
    print("error: Invalid operator is selected!")
    
    
```

## Program Explanation

This program is a simple calculator built using Python.  
It allows the user to enter two numbers and select an arithmetic operation.

1. The program uses `input()` to take values from the user.
2.  The inputs are converted to float so decimal numbers can be used.
3.  Conditional statements (`if`, `elif`, `else`) are used to decide which
  operation to perform.
4.  The program handles division by zero and invalid operators.
5.  Clear messages are shown for both results and error
